#Q1: Rock, Paper, Scissors

This is a rock, paper, scissors game that allows users to interact with computer

```python
import random

def weightedDM(rock, paper, scissor):
    """
    calculates the probability of rock, paper and scissors being thrown. 
    All of the probabilites start off at 33% (equal chance of being played). 
    If the player is playing more scissors, rock will be thrown and if they are playing more rock, paper will be thrown. 
    The probability of rock and paper being thrown must be within 10% for this function to trigger. 
    By game theory, paper is a smarter move since the worst result would be a draw. 
    
    """
    total = rock + paper + scissor
    rockProb = int(float(scissor) / float(total) * 100) 
    paperProb = int(float(rock) / float(total) * 100) + rockProb 
    

    rockStat = float(rock) / total
    paperStat = float(paper) / total
    scissorStat = float(scissor) / total
    
    dice = random.randint(1, 100)
    
    if (abs(float(rockStat) - float(paperStat))) <= .1 and (scissorStat <= .2):
        compThrow = 2
    elif dice <= rockProb:
        compThrow = 1
    elif dice >= rockProb and dice <= paperProb:
        compThrow = 2
    else:
        compThrow = 3
   
    return compThrow
```

```python
def outcome(playerThrow, compThrow):
    """
    This function first figures out what each player's move is and prints the result. 
    The function then takes the player's move and the computer's move and compares them to determine the outcome of the game. 
    
    """
    print ""
    if playerThrow == 1:
        print "Player threw ROCK!"
        playerValue = 3
    elif playerThrow == 2:
        print "Player threw PAPER!"
        playerValue = 1
    else:
        print "Player threw SCISSORS!"
        playerValue = 2
        
    if compThrow == 1:
        print "Computer threw ROCK!"
        compValue = 3
    elif compThrow == 2:
        print "Computer threw PAPER!"
        compValue = 1
    else:
        print "Computer threw SCISSOR!"
        compValue = 2
        
    print ""
        
    if playerValue == 1 and compValue == 3:
        compValue = compValue - 3
    elif playerValue == 3 and compValue == 1:
        playerValue = playerValue - 3
        
    if playerValue == compValue:
        print "Draw!"
        return 0
    elif playerValue > compValue:
        print "Player Wins!"
        return 1
    elif playerValue < compValue:
        print "Player Loses!"
        return 0
```

```python
def splitQuote(fullquote):
    """
    This function splits the quotes into words and returns a list list of all the quotes
    """
    import re  
    return re.split('\W+', fullquote)
```

```python
def wordsFromFullQuote(fullquote):
    """
    wordsFromFullQuotes generates a word list from a single quote. 
    It first splits the quotes in to words and then ignores empty words and makes all of the words lowercase. 
    The function adds each word to the list wordlist[] and returns the list. 
    """
    wordlist = []
    for w in splitQuote(fullquote): 
        if w: 
            wordlist.append(w.lower()) 
    return wordlist
```

```python
def  dictionaryFromAllQuotes():
    """
    This function  generates a posting list dictionary, whose keys are full quotes and whose values are dictionaries
    with the key being the word and the value being the number of times the word occurs in the full quote. 
    The function splits the quote into words and if the word has occured 1 is added to its count. 
    If the words has not occured, its value is set to 1. 
    """
    pldict = {}
    plsingledict = {}
    for fullquote in allquotes[0:3]: 
        wordlist = wordsFromFullQuote(fullquote) 
        for word in wordlist: 
            if word in plsingledict: 
                plsingledict[word] = plsingledict[word] + 1 
            else: 
                plsingledict[word] = 1 
        pldict[fullquote] = plsingledict 
        plsingledict = {} 
    return pldict 
```

```python
def reverseDictionaryFromAllQuotes():
    """
    This function generates a reverse posting-list dictionary. 
    The keys in the dictionary are the words and the values are dictioniares with the key being a full quote
        and the value being the number of times the word appeared in the full quote. 
    
    For each quote in the full quote list, the quote is split into words and for each word, if it is not in the word list yet, it is added to fullwordlist[].
    
    The function then goes through each word in fullwordlist[], the function searches the quotes in the file. 
    if the word is in the quote, the word count is stored in the dictionary, that is then stores in the post-listing dictionary. 
    """
    rpldict = {}

    fullwordlist = [] 
    for fullquote in allquotes: 
        wordlist = splitQuote(fullquote) 
        for word in wordlist: 
            if word not in fullwordlist: 
                if word: 
                    fullwordlist.append(word) 
    

    rplsingledict = {} 
    for word in fullwordlist: 
        for fullquote in allquotes: 
            if word in fullquote: 
                rplsingledict[fullquote] = fullquote.count(word) 
        rpldict[word] = rplsingledict
        rplsingledict = {} 

    return rpldict
```

```python
def getTf(word, quote):
    """
    This function gets the TF of a word/quote. 
    IT does this by getting the count of the most common word. 
    Term frequency = occurences of the word / occurences of the most common word. 
    """
    quote = splitQuote(quote) 
    while '' in quote: 
        quote.remove('')
        

    wordcount = []
    for word in quote:
        wordcount.append(quote.count(word))
    return quote.count(word) / float(max(wordcount)) 
```

```python
def getIdf(word):
    
    """
    This function gets the IDF of a word. 
    It does this by counting the number of quotes containing word and then taking the 
        log(number of quotes / number of quotes containing this word)
    """
    import math
    
    x = 1
    for quote in allquotes:
        if word in splitQuote(quote):
            x += 1
    return math.log((len(allquotes) / x), 10) 
```

```python
def getTfIdf(word, quote):
    """
    This function gets the TF-IDF of the word by multiplying the TF by the IDF
    """
    
    return getTf(word, quote) * getIdf(word)
```

```python

def searchByWord(word):
    """
    This function searches quotes using a single word. 
    It does this by taking a word as an argument and returning a dictionary 
        whose keys are full quotes containing one or more of the words in that list 
        and whose values are the sum of TF-IDF scores of the words. 
    """
    results = {}
    for quote in allquotes:
        if word in splitQuote(quote):
            results[quote] = getTfIdf(word, quote)
    return results
```

```python
def searchByWordList(wordlist):
    """
    This function searches quotes using a list of words as an argument adn returning a dictionary 
        whoses keys are full quotes containing one or more of the words in the list and whose values are the sum of the TF-IDF scores. 
    """
    results = {}
    for quote in allquotes:
        for word in wordlist:
            if word in splitQuote(quote):
                if quote in results:
                    results[quote] = results[quote] + getTfIdf(quote, word)
                else:
                    results[quote] = getTfIdf(quote, word)
    return results
```
